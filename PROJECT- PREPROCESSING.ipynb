## Data Preparation

```python
#dropping duplicate column
data.drop("ua", axis=1, inplace=True)
```

```python
#data.rename(columns={"ub":"area_name"}, inplace= True)
```

```python
# dropping duplicates
data = data.drop(labels=[151,1267,1266], axis=0)
data.shape
```

```python
data.fillna(value=0,inplace=True)
```

```python
data=data[~((data["Republican"]==0)&(data["Democrat"]==0))]
```

```python
data.shape
```

```python
#creating a column 'winner', '0' and '1' represents winners democrat and Republican respectively
data["winner"]=np.where(data['Democrat'] > data['Republican'], 0, 1)
```
